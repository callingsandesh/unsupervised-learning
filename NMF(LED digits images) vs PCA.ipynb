```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import csv
```

```python
data_file = open('lcd-digits.csv')
led_digits = csv.reader(data_file , delimiter = ',')
samples=[]
for item in led_digits:
    rows=[]
    for i in item:
        rows.append(float(i))
    samples.append(rows)
samples=np.array(samples)
```

## function to display image

```python
def show_as_image(sample):
    bitmap = sample.reshape((13, 8))
    plt.figure()
    plt.imshow(bitmap, cmap='gray', interpolation='nearest')
    plt.colorbar()
    plt.show()
```

```python
show_as_image(samples[10])
```

```python
for i in samples:
    show_as_image(i)
```

```python
# Import NMF
from sklearn.decomposition import NMF

# Create an NMF model: model
model = NMF(n_components = 7)

# Apply fit_transform to samples: features
features = model.fit_transform(samples)

# Call show_as_image on each component
for component in model.components_:
    show_as_image(component)

# Assign the 0th row of features: digit_features
digit_features = features[99]

# Print digit_features
print(digit_features)
```

Take a moment to look through the plots and notice how NMF has expressed the digit as a sum of the components!

### PCA doesn't learn parts

```python
# Import PCA
from sklearn.decomposition import PCA

# Create a PCA instance: model
model = PCA(n_components = 7)

# Apply fit_transform to samples: features
features = model.fit_transform(samples)

# Call show_as_image on each component
for component in model.components_:
    show_as_image(component)
```

Notice that the components of PCA do not represent meaningful parts of images of LED digits!
